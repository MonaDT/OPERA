```python
import pandas as pd
import matplotlib.pyplot
import matplotlib.dates

from datetime import datetime

df = pd.read_csv("../Materials/KVO_META.csv", encoding='latin-1', index_col=False)
```

```python
#De Datum kolom wordt omgezet naar dtype datetime.

df['Datum'] = df['Datum'].str.strip()
df['Datum'] = pd.to_datetime(df['Datum'], format="%d/%m/%Y")

#blanco pagina's worden verwijderd

old_len = len(df)
df = df.dropna(subset=['Productie','Datum'])
print(f'{old_len - len(df)} of {old_len} records removed')
```

```python
def bar_chart_maker (column, n_bars = 3, period_length = 2000):
    
    output = pd.pivot_table(df, index= ['Season'], values = ['Production'], columns = [column], aggfunc='count')
    bar_chart = bcr.bar_chart_race(df=output, filename=None, period_length=period_length, title=str(column), n_bars=3, fixed_max=True, filter_column_colors = True)
    return bar_chart

```
