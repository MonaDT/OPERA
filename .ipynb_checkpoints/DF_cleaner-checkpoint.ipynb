# You can just run the cells below. they're simple preprocessing steps

```python
import pandas as pd

df = pd.read_excel(r'C:\Users\monaa\Dropbox\Thesis\Primair\Metadata_OBV_gesplitst.xlsx')
df.head()
```

```python
import random
df['Voorstellings_ID'] = random.sample(range(100000, 999999), df.shape[0])
```

```python
#indien er geen productie en datum is, gaat het om een blanco of minder relevante pagina (bv. advertenties)
#in dat geval wordt de rij uit de df verwijderd

old_len = len(df)
df = df.dropna(subset=['Productie','Datum'])
print(f'{old_len - len(df)} of {old_len} records removed')
```

```python
#Verwijder de voorstelling indien de productie én datum dubbel in df staan

all_dups = len(df)

df = df.drop_duplicates(subset=['Productie', 'Datum'])
print(f'{all_dups - len(df)} of {all_dups} records removed')

len(df)
```

```python
#Boolean column voor aant te geven of het om een muktiple bill gaat

multiple_bills = []
for row in df['Productie']:
    if '+' in row:
        
        multiple_bills.append(True)
    else:
        multiple_bills.append(False)


df['Multiple bill'] = multiple_bills
df['Multiple bill'].value_counts()
```

```python
df = df.assign(Productie=df.Productie.str.split("+")).explode('Productie')
df = df.reset_index(drop=True)
df.head()
```

```python
df['Productie'] = df['Productie'].str.strip()
```

```python
#attempt number two. This kindof works, but because it's a nested for-loop it will add every title twice

clean_title = []

for row in df['Productie']:
    for key, values in variatie.items():
        if row in values:
            clean_title.append(key)
        else:
            clean_title.append(row)
```

```python
df['Datum'].replace(to_replace=["?", "? "], value = "01/01/2100", inplace = True)
df['Datum'].replace(to_replace=["11/11923"], value = "11/01/1923", inplace = True)
df['Datum'].replace(to_replace=["13/21897"], value = "13/02/1923", inplace = True)
df['Datum'].replace(to_replace=["21/1/1012"], value = "21/1/1912", inplace = True)
df['Datum'].replace(to_replace=["30/1/1013"], value = "30/1/1913", inplace = True)
df['Datum'].replace(to_replace=["27/2/191"], value = "27/2/1919", inplace = True)
df['Datum'].replace(to_replace=["15/4/1026"], value = "15/4/1926", inplace = True)
df['Datum'].replace(to_replace=[" 27/10/1894"], value = "27/10/1894", inplace = True)

df['Datum'].replace(to_replace=["1900/01/02"], value = "02/01/1900", inplace = True)
df['Datum'].replace(to_replace=["25/12/2909"], value = "25/12/1909", inplace = True)
```

```python
cleaned_dates = []
for row in df['Datum']:
    row = str(row).strip()
    
    if '-' in row:
        row = row.replace('-', '/')
        
    if '00:00:00' in row:
        
        row = row.replace('00:00:00', '')
    cleaned_dates.append(row)
```

```python
df['Datum_cleaned'] = cleaned_dates
df['Datum_cleaned'] = df['Datum_cleaned'].astype(str)
df['Datum_cleaned'] = df['Datum_cleaned'].str.strip()
```

```python
#df['Datum_cleaned'] = pd.to_datetime(df['Datum_cleaned'], format = '%d/%m/%Y')
df['Datum_cleaned'] = pd.to_datetime(df['Datum_cleaned'])
```
