# Bar chart race

```python
import pandas as pd
import matplotlib.pyplot
import matplotlib.dates
import bar_chart_race as bcr
import numpy as np
from datetime import datetime

#import df with all metadata.
df = pd.read_csv("../Materials/KVO_META.csv", encoding='latin-1', index_col=False)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#The date column should be transferred to datetime
df['Date'] = df['Date'].str.strip()
#df['Date'] = pd.to_datetime(df['Date'], format="%d/%m/%Y")
```

```python
#create a pivot table that keeps track how often each language occurs per season
output = pd.pivot_table(df, index= ['Season'], values = ['Production'], columns = ['Original language'], aggfunc='count')

#Create bar chart which keeps track of the most common original languages per season.

bar_chart_lang = bcr.bar_chart_race(df=output, 
                               filename=None, 
                               period_length=2000, 
                               title='Original language', 
                               n_bars=3, 
                               fixed_max=True, 
                               filter_column_colors = True)
```
